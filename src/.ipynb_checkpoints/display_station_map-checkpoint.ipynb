```python
import os
from mpl_toolkits.basemap import Basemap
import numpy as np
import csv
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
station_file = 'D:\\Data\\全国空气质量\\站点列表-2019.08.01起.csv'
img1File = 'C:\\Users\\zpyin\\Desktop\\Wuhan AOD analysis\\img\\station_map.png'
img2File = 'C:\\Users\\zpyin\\Desktop\\Wuhan AOD analysis\\img\\station_map_fine.png'
```

Read the station list

```python
station = []
with open(station_file, 'r', newline='', encoding='utf-8') as fh:
    data = csv.DictReader(fh, fieldnames=['ID', 'name', 'city', 'lon', 'lat'], delimiter=',')
    for row in data:
        station.append(row)

station = station[1:]
```

Data visualization

```python
plt.figure(figsize=[12,8], linewidth=3, dpi=300)
m = Basemap(llcrnrlon=77,llcrnrlat=14.0,urcrnrlon=140.,urcrnrlat=51.0, projection='lcc', 
            resolution='l', lat_1=33, lat_2=45, lon_0=100)
m.readshapefile("D:\\Data\\china_boundary_shp\\gadm36_CHN_shp\\gadm36_CHN_0", 'states', drawbounds=True, linewidth=3)
m.readshapefile("D:\\Data\\china_boundary_shp\\gadm36_CHN_shp\\gadm36_CHN_1", 'states', drawbounds=True)
m.readshapefile('D:\\Data\\china_boundary_shp\\gadm36_TWN_shp\\gadm36_TWN_1', 'states', drawbounds=True, linewidth=3)

# m.drawcoastlines()
# m.drawcountries()
# m.drawmapboundary(linewidth=3)
m.etopo()

for station_info in station:
    if station_info['lon']:
        x, y = m(float(station_info['lon']), float(station_info['lat']))
        if station_info['city'] == u'武汉':
            m.scatter(x, y, s=30, marker='o', edgecolor='k', facecolor='r')
        else:
            m.scatter(x, y, s=15, marker='o', edgecolor='b', facecolor='b')
    
plt.title('National ground stations map', fontsize=25, fontname='Times New Roman')
plt.show()

plt.savefig(img1File, dpi=300)
```
